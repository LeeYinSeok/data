```python
# cv2, numpy, matplotlib.pyplot
import cv2 # 컴퓨터비전용 라이브러리
import numpy as np # 행렬 연산
import matplotlib.pyplot as plt #시각화 라이브러리
```

```python
img = cv2.imread("image/Lenna.png") # BGR -> RGB

plt.imshow(img)
```

```python
img = cv2.imread("image/Lenna.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
```

```python
# add(100), subtract(100), multiply(1.7), divide(3)
ex1_result1 = cv2.add(img, 100)
ex1_result2 = cv2.subtract(img, 100)
ex1_result3 = cv2.multiply(img, 1.7)
ex1_result4 = cv2.divide(img, 3)
```

```python
#matplotlib.pyplot -> subplot(2.3), imshow()
plt.subplot(2,3,1); plt.imshow(img, cmap='gray')
plt.subplot(2,3,2); plt.imshow(ex1_result1, cmap='gray')
plt.subplot(2,3,3); plt.imshow(ex1_result2, cmap='gray')
plt.subplot(2,3,4); plt.imshow(ex1_result3, cmap='gray')
plt.subplot(2,3,5); plt.imshow(ex1_result4, cmap='gray')
```

```python
#matplotlib.pyplot -> subplot(2.3), imshow()
plt.subplot(2,3,1); plt.imshow(img, cmap='gray')
plt.axis('off'); plt.gca().set_title('original')
plt.subplot(2,3,2); plt.imshow(ex1_result1, cmap='gray')
plt.axis('off'); plt.gca().set_title('add')
plt.subplot(2,3,3); plt.imshow(ex1_result2, cmap='gray')
plt.axis('off'); plt.gca().set_title('subtract')
plt.subplot(2,3,4); plt.imshow(ex1_result3, cmap='gray')
plt.axis('off'); plt.gca().set_title('multiply')
plt.subplot(2,3,5); plt.imshow(ex1_result4, cmap='gray')
plt.axis('off'); plt.gca().set_title('divide')
```
